```python
# Convert to numeric (coerce errors will turn invalid entries into NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
print(df['TotalCharges'].isnull().sum())
```

```python
df['TotalCharges'].fillna(0, inplace=True)
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
binary_cols = ['Partner','Dependents','PhoneService','PaperlessBilling','Churn']
```

```python
for col in binary_cols:
    df[col] = df[col].map({'Yes':1, 'No':0})
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].map({0:'No', 1:'Yes'})
```

```python
print("Null values after cleaning:")
print(df.isnull().sum())
```

```python
print("Unique values in Churn:", df['Churn'].unique())
print("Unique values in Contract:", df['Contract'].unique())
print("Unique values in PaymentMethod:", df['PaymentMethod'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set style
sns.set(style="whitegrid")
```

```python
# 1. Overall Churn Rate
churn_rate = df['Churn'].mean() * 100
print(f"Overall Churn Rate: {churn_rate:.2f}%")
```

```python
# 2. Churn Count Plot
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', hue='Churn', data=df, palette="Set2", legend=False)
plt.title("Churn Distribution (0 = No, 1 = Yes)")
plt.show()
```

```python
# 3. Churn by Gender
plt.figure(figsize=(6,4))
sns.countplot(x='gender', hue='Churn', data=df, palette="Set1")
plt.title("Churn by Gender")
plt.show()
```

```python
# 4. Churn by Senior Citizen
plt.figure(figsize=(6,4))
sns.countplot(x='SeniorCitizen', hue='Churn', data=df, palette="coolwarm")
plt.title("Churn by Senior Citizen")
plt.show()
```

```python
# 5. Churn by Contract Type
plt.figure(figsize=(6,4))
sns.countplot(x='Contract', hue='Churn', data=df, palette="Paired")
plt.title("Churn by Contract Type")
plt.show()
```

```python
# 6. Monthly Charges Distribution
plt.figure(figsize=(6,4))
sns.histplot(df['MonthlyCharges'], bins=30, kde=True, color="black")
plt.title("Distribution of Monthly Charges")
plt.show()
```

```python
# 7. Heatmap for Correlation (numerical features)
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="Blues", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
